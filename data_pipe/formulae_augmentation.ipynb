Варианты аугментации:
1) словарь синонимов
2) моделирование опечаток
4) случайное удаление спец. символов
5) комбинация


```python
import re
import random
from random import randint
```

```python
# моделирование сложных команд
# измененное регулярное выражение для поиска выражения вида \frac{...}{...}
complex_pattern = r'\\frac{([^{}]+)}{([^{}]+)}|\\int_{([^{}]+)}\^{([^{}]+)}|\\lim_{([^{}]+)}'

# Функция для замены найденных выражений
def replace_expression(match):
    if match.group(1):
        frac_list = [f"дробь от {match.group(1)} делить на {match.group(2)}", f"{match.group(1)} делить на {match.group(2)}",
            f"{match.group(1)} разделить на {match.group(2)}"]
        return random.choice(frac_list)
    elif match.group(3):
        return f"определённый интеграл от {match.group(3)} до {match.group(4)} по "
    elif match.group(5):
        return f"предел при {match.group(5)} от"

# пример использования для замены 
# modified_text = re.sub(complex_pattern, replace_expression, latex_text)
```

```python
# для простых синонимов
def replace_with_synonyms(inp_str, replacement_dict, probability):
    # пояснение - собираем нашу регулярку
    # для ЕДИНИЧНЫХ БУКВ - смотрим что у них нет слева и справа \\ или других букв
    
    pattern = '|'.join(rf'(?<![a-zA-Z\\])({re.escape(key)})(?![a-zA-Z])' if key.isalpha() else 
                      rf'(?<!\\)({re.escape(key)})(?![a-zA-Z])'
                       for key in replacement_dict)
   
    def replacer(match):
        key = match.group(0)
        for repl_key, repl_value in replacement_dict.items():
            if random.random() < probability and key == repl_key:
                replacement = random.choice(repl_value)
                return f' {replacement} '
        return key  


    result = re.sub(pattern, replacer, inp_str)
    return result


frms_synonyms = []
replacement_prob = 0.6

```

```python
# случайное удаление $, \, \\, [], (), {}, |, ||, +, -, =, :,* и т.д.
 
def delete_symbols(inp, prob):
    symbols_to_delete = ['$','\\','[',']','(',')','{','}','|',':',';',"&",'*','#','^','_']
    
    for symb in symbols_to_delete:
        if random.random() < prob:
            inp=inp.replace(symb,' ', randint(1,len(inp)//2))
    return inp

deletion_prob = 0.05
```

```python
# для рандомизации буков
cyr_letters="АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯабвгдеёжзиклмнопрстуфхцчшщъыьэюя"
lat_letters="ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"
```

```python
# для использования typo только на не-числах
def make_typos(input_string):
    res = ""
    # проблема с каждой десятой буквой (чтобы потом было что заменять по синонимам)
    prob = 0.05
    for i, c in enumerate(input_string):
        # если это не число и не начало слова, то с вероятностью prob мы что-то делаем
        if  random.random() < prob and not c.isnumeric() and \
               ( (c.isalpha() and input_string[i-1].isalpha()) or (not c.isalpha()) ):
            # вероятность вообще как-либо что-либо менять
            op = randint(1,4)
            if c!=' ':
                # повторяем
                if op==1:
                    res+=c*2 
                # добавляем рандомный пробел
                if op==2:
                    res+=f'{c} '
                # меняем на рандомную букву
                if op==3:
                    # если это буква латинского алфавита
                    if c in lat_letters:
                        res+=random.choice(lat_letters)
                    # в других случаях
                    else:
                        res+=random.choice(cyr_letters)
                # ничего не делаем
                if op==4:
                    res+=c
            else:
                # рандомно удаляем пробелы
                if op==1 or op==3:
                    res+=''
        else:
            res+=c
    return res
```

```python
# сначала - сложные синонимы
# потом - простые
# потом удаление спец. символов
# и напоследок - порча букв
```

```python
# основной цикл аугментации
# собираем три колонки
# рандомные удаления и ошибки
# все виды аугментации сразу
# и только синонимы

# собираем плоский словарь из всех остальных словарей
# сначала - всё в один лист
all_items = [(key, value) for d in replacement_dicts for key, value in d.items()]
# сортируем по ключам - от самого длинного к самому короткому
replacement_dict = {key: value for key, value in sorted(all_items, key=lambda x: len(x[0]), reverse=True)}

# для случайных удалений
frms_stripped = []
# для всего сразу
frms_all_augments = []
# для синонимов
frms_syns = []

i=0
for formula in tqdm(df['original_formula']):
    # аугментируем сложными заменами
    comp_pats = re.sub(complex_pattern, replace_expression, fr'{formula}')
    # аугментируем простыми заменами
    simp_pats = replace_with_synonyms(comp_pats, replacement_dict, replacement_prob)
    # добавляем это в лист для чисто синонимичных замен
    frms_syns.append(simp_pats)
    
    # теперь аугментируем случайными удалениями
    # при этом тут будет два раза потому что у нас ещё есть колонка чисто с ними
    
    # без синонимов
    stripped_only = delete_symbols(formula, deletion_prob)
    # с синонимами
    stripped_syn = delete_symbols(simp_pats, deletion_prob)
    
    # предположим что слишком короткие формулы смысла аугментировать через опечатки нет
    # т.к. удаление спец. символов уже их достаточно аугментирует
    if len(formula)>30:
        # это собственно и будут финальные варианты
        # print("stripping done")
        stripped_only = make_typos(stripped_only)
        stripped_syn = make_typos(stripped_syn)
        
    frms_all_augments.append(stripped_syn)
    frms_stripped.append(stripped_only)
    
    if i<10:
        print(f"\noriginal: {formula}\nall_synonyms:{simp_pats}\nstripped_only:{stripped_only}\nall_augments:{stripped_syn}")
    i+=1
```

```python
# sanity check
if len(frms_stripped) == len(frms_all_augments)==len(frms_syns)==len(df['original_formula']):
    print("OK!!")
```
